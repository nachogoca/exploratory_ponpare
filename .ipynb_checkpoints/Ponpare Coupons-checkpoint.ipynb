## What are the most common  offered coupons' genre?

```python
# Get coupon data
coupons = pd.read_csv('../data/coupon_list_train_translated.csv')
```

```python
coupons.head()
```

```python
# Delete columns I won't use
coupons.drop(['DISPFROM', 'DISPEND','DISPPERIOD','VALIDFROM','VALIDEND','VALIDPERIOD','USABLE_DATE_MON','USABLE_DATE_TUE','USABLE_DATE_WED','USABLE_DATE_THU','USABLE_DATE_FRI','USABLE_DATE_SAT','USABLE_DATE_SUN','USABLE_DATE_HOLIDAY','USABLE_DATE_BEFORE_HOLIDAY','large_area_name','ken_name','small_area_name'], axis=1, inplace=True)
```

```python
# Offered coupons
coupons_by_genre = pd.DataFrame(coupons.groupby('GENRE_NAME').count()['COUPON_ID_hash'])
coupons_by_genre.columns = ['Count']
coupons_by_genre.sort_values(['Count'], ascending=False, inplace=True)
coupons_by_genre.plot(kind='bar')
plt.title("Offered coupons")
plt.show()
```

### Let's join the information about users and coupons with the logs

```python
# Merge with coupons
coupon_total = pd.merge(left=coupon_visit_with_users, right=coupons, left_on='VIEW_COUPON_ID_hash', right_on='COUPON_ID_hash')
```

```python
coupon_total.drop(['PAGE_SERIAL','PURCHASEID_hash','COUPON_ID_hash'], axis=1, inplace=True)
```

```python
coupon_total.head()
```

```python
coupon_seen_count = coupon_total.groupby('GENRE_NAME')['PURCHASE_FLG'].count()
coupon_seen_count.sort_values(inplace=True, ascending=False)
coupon_seen_count.plot(kind='bar')
plt.title('Coupons seen by genre')
plt.show()
```

```python
coupon_seen_sum = coupon_total.groupby('GENRE_NAME')['PURCHASE_FLG'].sum()
coupon_seen_sum.sort_values(inplace=True, ascending=False)
coupon_seen_sum.plot(kind='bar')
plt.title('Coupons bought by genre')
plt.show()
```

## Is there a correlation between gender/age and genre? 

```python
coupon_seen_count_by_sex = coupon_total.groupby(['GENRE_NAME', 'SEX_ID'])['PURCHASE_FLG'].count()
to_plot = coupon_seen_count_by_sex.unstack().sort_values(['m','f'], ascending=False)
to_plot.plot(kind='bar')
plt.ylabel('Coupons seen')
plt.title('Coupons seen by sex')
plt.show()
```

The most frequent coupon genre seen by men are Food and Hotels, where they see more coupons than women. But in the genre of Delivery Service women are by far the most frequent user.

```python
coupon_bou_count_by_sex = coupon_total.groupby(['GENRE_NAME', 'SEX_ID'])['PURCHASE_FLG'].sum()
to_plot = coupon_bou_count_by_sex.unstack().sort_values(['m','f'], ascending=False)
to_plot.plot(kind='bar')
plt.ylabel('Coupons bought')
plt.title('Coupons bought by sex')
plt.show()
```

The interesting thing is that men don't actually buy the hotels they see. Although they do buy the food, at a bigger rate than women. Delivery service is still the most frequent genre, dominated by feminine users.

```python
coupon_seen_count_by_age_and_sex = coupon_total.groupby(['GENRE_NAME', 'SEX_ID', 'AGE'])['PURCHASE_FLG'].count()
```

```python
# Get count by age and sex of genres bought
to_plot = coupon_seen_count_by_age_and_sex.fillna(0)
p = pd.DataFrame(to_plot)

# Getting ready to plot
genres = coupons['GENRE_NAME'].unique()
sexs = ['m','f']
colors = ['blue', 'red']
f, axarr = plt.subplots(len(genres) * 2, 1, figsize=(10,100))
f.subplots_adjust(hspace=0.6)

# Two subplots per genre
for genre_index in range(len(genres)):
    # Get subdata of selected genre
    subdata = p.xs(genres[genre_index])
    y_lim = subdata.values.max()
    
    # One subplot for genre
    for i in range(2):
        x = subdata.xs(sexs[i]).index.values
        y = subdata.xs(sexs[i]).values
        
        subplt = axarr[(genre_index*2)  + i]
        subplt.bar(x, y, color=colors[i])
        
        # Labels and stuff
        subplt.set_xlim([10,80])
        subplt.set_ylim([0,y_lim])
        subplt.set_xlabel("Age")
        subplt.set_ylabel("Coupons seen")
        
        subplt.set_title('Coupons seen by Genre: ' + genres[genre_index] + ',   Sex: ' + sexs[i])
```

#### Observations:

The genres Hair salon, spa, relaxation, beauty and Nail and eye salon are dominated by women of all ages. Except Beauty, where middle-aged women are the dominant users.

In the genres Food, Delivery service, Leisure, Hotel and Japanese hotel there is a shift to older ages, especially in hotels for men.

## Let's try to predict the sex of the user.
This is useful in case of an anonymous session. Where your offer can be more focused to the user, if you know his/her gender.

### Dataset preparation

```python
users_total_activity = users.join(user_activity, on='USER_ID_hash')
users_total_activity.drop('USER_ID_hash', axis=1, inplace=True)
users_total_activity.head()
```

```python
# Zero -> Masculine
# One -> Feminine
users_total_activity.replace(['m','f'],[0,1], inplace=True)
```

```python
users_total_activity.dropna(inplace=True)
```

### Nomalization

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
np_scaled = min_max_scaler.fit_transform(users_total_activity)
users_total_activity_normal = pd.DataFrame(np_scaled)
```

```python
users_total_activity_normal.columns = users_total_activity.columns
```

```python
users_total_activity_normal.describe()
```

```python
users_total_activity_normal.head()
```

### Separate dataset into train and test datasets

```python
from sklearn.model_selection import train_test_split

y = users_total_activity_normal['SEX_ID']
X = users_total_activity_normal.drop('SEX_ID', axis=1).as_matrix()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=45)
```

## Supervised learning with Support Vector Machines

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(X_train,y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf.predict([.15, 0.2, 0.2, 0.1, 0.1, 0.3, 0.0, 0.0, 0.0, 0.1, 0.2, 0.2, 0.2, 0.3])
```

### Linear regression to predict the age


```python
from sklearn.model_selection import train_test_split
from sklearn import svm

y = users_total_activity['AGE'].as_matrix()
X = users_total_activity_normal.drop('AGE', axis=1).as_matrix()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=45)

clf = svm.SVR()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```
